```python
import os
import re
import json

def get_values(output_folder, metric, search_string):
    values_dict = {}

    for file_name in os.listdir(output_folder):
        if file_name.endswith('_node2vec_log_GAT.txt'):
            file_path = os.path.join(output_folder, file_name)
            chromosome_name = file_name.split('_')[1]  

            with open(file_path, 'r') as file:
                lines = file.readlines()

                for line in lines:
                    if search_string in line:
                        value = float(line.split(': ')[1].strip())

                        values_dict[chromosome_name] = value
                        break

    return values_dict

def sort_key(chromosome):
    """Sort chromosomes with numeric and handle 'X' at the end."""
    match = re.match(r'chr(\d+)', chromosome)
    if match:
        return int(match.group(1))  
    elif chromosome == 'chrX':
        return 23  
    return float('inf'), chromosome  

def save_sorted_values(values_dict, output_file):
    sorted_values = dict(sorted(values_dict.items(), key=lambda x: sort_key(x[0])))

    with open(output_file, 'w') as f:
        json.dump(sorted_values, f, indent=4)

    print(f'Sorted values saved to {output_file}')

output_folder = 'Outputs/node2vec_GAT_reduced_GM12878'

dSCC_values = get_values(output_folder, metric="dSCC", search_string="Optimal dSCC")
MSE_values = get_values(output_folder, metric="MSE", search_string="Final MSE loss")

output_file_dSCC = 'Outputs/node2vec_GAT_reduced_GM12878/sorted_dSCC_GAT_values.json'
save_sorted_values(dSCC_values, output_file_dSCC)

output_file_MSE = 'Outputs/node2vec_GAT_reduced_GM12878/sorted_MSE_GAT_values.json'
save_sorted_values(MSE_values, output_file_MSE)

print("dSCC values:", dSCC_values)
print("MSE values:", MSE_values)
```

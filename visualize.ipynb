```python
def count_residues(pdb_filename):
    with open(pdb_filename, 'r') as pdb_file:
        residues = set()  

        for line in pdb_file:
            if line.startswith('ATOM') or line.startswith('HETATM'):
                res_id = line[22:26].strip() 
                residues.add(res_id)  #

        num_residues = len(residues)
        return num_residues
 
pdb_filename = 'Outputs/CH12-LX_chr12_1mb_GATSmallerNet_LINE_structure.pdb'
num_residues = count_residues(pdb_filename)
print(f'Number of residues in the PDB file: {num_residues}')
```

```python
import matplotlib.pyplot as plt
import numpy as np

pdb_filename = 'Outputs/CH12-LX_chr12_1mb_SmallerNet_LINE_structure.pdb'
with open(pdb_filename, 'r') as pdb_file:
    pdb_data = pdb_file.read()

view = py3Dmol.view(width=800, height=600)
view.addModel(pdb_data, 'pdb')

num_residues = 119
colors = plt.cm.get_cmap('viridis', num_residues)  

for i in range(1, num_residues + 1): 
    color = colors(i / num_residues)  
    hex_color = '#%02x%02x%02x' % tuple([int(c * 255) for c in color[:3]])  
    view.setStyle({'resi': i}, {'stick': {'color': hex_color}})

view.zoomTo()
view.show()
```
